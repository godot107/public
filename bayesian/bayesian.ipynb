```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_csv('./data/drug200.csv')
```

```python
data.head()
```

```python
pt = pd.pivot_table(data, index = 'Sex', columns = 'Cholesterol', aggfunc = 'count', values = 'Age')
pt
```

```python
pop = np.sum(pt.values)
```

Given Male, what is the probability Cholesterol is 'High'?

$$ P(High| Male)= \frac {P(High)P(Male | High)}{P(Male)}$$

```python
#

p_male = (pt.loc['M', 'HIGH'] + pt.loc['M', 'NORMAL']) / pop

p_high =  (pt.loc['F', 'HIGH'] + pt.loc['M', 'HIGH']) / pop

p_male_high = pt.loc['M', 'HIGH']/ (pt.loc['F', 'HIGH'] + pt.loc['M', 'HIGH'])


print(f"P(High) = {p_high:.2f}")
print(f"P(Male) = {p_male}")
print(f"P(Male|High) = {p_male_high:.2f}")
```

```python
p_high_male_1 = p_high * p_male_high / p_male

p_high_male_2 = pt.loc['M', 'HIGH']/ (pt.loc['M', 'HIGH'] + pt.loc['M', 'NORMAL'])
```

```python
p_high_male_1
```

```python
p_high_male_1 == p_high_male_2
```
