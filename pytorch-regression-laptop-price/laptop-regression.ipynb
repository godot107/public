## Model Architecture

```python


class MLP(nn.Module):
    def __init__(self,num_inputs= 1,num_outputs=1):
        super(MLP, self).__init__()

        self.fc1   = nn.Linear(in_features=num_inputs, out_features= 24)
        self.relu = nn.ReLU()        
        self.fc2   = nn.Linear(in_features= 24, out_features= 12)
        self.fc3   = nn.Linear(in_features= 12, out_features= 6)

        
        self.output   = nn.Linear(in_features= 6, out_features= num_outputs) 
        
    def forward(self, X):

        # 1st layer
        X = self.fc1(X)
        X = self.relu(X)
        
        #2nd layer
        X = self.fc2(X)
        X = self.relu(X)      
        #3rd layer
        X = self.fc3(X)
        X = self.relu(X)
        
        #output layer
        X = self.output(X)
        
        return X
        
```

```python
# training parameters
n_epochs = 50   # number of epochs to run
loss_function = nn.MSELoss()

# lrs=[5e-2, 1e-1, 5e-1, 0.01, 1e-3]
lrs=[0.001]
```

```python
loss_ls = []
val_loss_ls = []
epochs = []
```

```python
import torch.optim as optim

# model = MLP()
# optimizer = optim.Adam(model.parameters(), lr=lr)

best_mse = np.inf
best_model = None
# label = []

for lr in lrs:
    model = MLP()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    for epoch in range(n_epochs): #

        model.train()
        for X, y in train_dl:


            # zero the parameter gradients
            optimizer.zero_grad()

            pred = model(X)
            loss = loss_function(pred, y)

            # Perform gradient descent

            loss.backward()
            optimizer.step()


        # current_best = eval_model(model,test_loader)
        # if epoch % 5 == 0:
        with torch.no_grad():
            model.eval()
            val_pred = model(X_val)
            val_loss = loss_function(val_pred, y_val)
            # print(val_loss.item())

            # Tracking loss Evaluation
            # label.append(lr)
            loss_ls.append(loss.item())
            val_loss_ls.append(val_loss.item())
            epochs.append(epoch)

            # store best model
            if val_loss < best_mse:
                best_model = model
                # print(lr)

```
