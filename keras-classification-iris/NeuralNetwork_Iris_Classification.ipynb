```python
# Import Libraries

from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import tensorflow as tf
from keras import models
from keras import layers
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
import pandas as pd
import numpy as np
from watermark import watermark

```

```python
print(tf.__version__)
```

```python
# Load DataFrame

df = load_iris(return_X_y = True, as_frame = True)
df = pd.concat((df[0], df[1]), axis = 1)
df = pd.concat((df, pd.get_dummies(df['target'], prefix = 'target')), axis = 1)
```

```python
sns.heatmap(df.corr(), annot = True)
```

```python
sns.pairplot(df.iloc[:,:-3], hue="target")
```

```python
feature_dim = 3

#split the dataset
X_train, X_test, y_train, y_test = train_test_split(df[['sepal length (cm)',
 'sepal width (cm)',
 'petal length (cm)']], df['target'], test_size=0.2, random_state = 42)

# One hot encoding
enc = OneHotEncoder(sparse_output = False)
y_train = enc.fit_transform(np.array(y_train).reshape(-1,1))
y_test = enc.transform(np.array(y_test).reshape(-1,1)) # Need to one-hot encode output for the neuralnetwork output layer

# Scaling
#scaler = preprocessing.StandardScaler()
#X_train_transformed = scaler.fit_transform(X_train)
#X_test_transformed = scaler.transform(X_test)

normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(X_train)
```

```python
enc.categories_
```

```python
load_iris()['target_names']
```

```python
# Building the Neural Network

network = models.Sequential()
#network.add(layers.Dense(16, activation = "relu", input_dim= feature_dim))
network.add(normalizer)
network.add(layers.Dropout(.2))
network.add(layers.Dense(16, activation = "relu"))
network.add(layers.Dropout(.2))
network.add(layers.Dense(units = 3, activation = 'softmax'))

# Compile Neural Network
network.compile(loss = "categorical_crossentropy",
                optimizer = "adam",
                metrics = ["accuracy"]
                )

network.summary()
```

```python
# Train Model
history = network.fit(X_train,#X_train_transformed,
                    y_train,
                    epochs = 50,
                    batch_size = 32,
                    validation_split = .33
                    #validation_data = (X_test_transformed, y_test)
)
```

```python
# Making Predictions with an array of feaures
predictions = network.predict(X_test)

# https://stackoverflow.com/questions/43589842/test-score-vs-test-accuracy-when-evaluating-model-using-keras
# categorical_crossentropy, accuracy
score = network.evaluate(X_test, y_test, verbose=0)
```

```python
score
```

```python
# Evaluating Loss:

training_loss = history.history["loss"]
test_loss = history.history["val_loss"]

epoch_count = range(1,len(training_loss) + 1)

plt.plot(epoch_count, training_loss, "r--")
plt.plot(epoch_count, test_loss, "b-")
plt.legend(["Training Loss","Test Loss"])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show();
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, threshold = roc_curve(y_test.ravel(), predictions.ravel())

plt.plot(fpr, tpr, label='AUC = {:.3f}'.format(auc(fpr, tpr)))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend();
```

```python
predictions
```

```python
prediction_decision_function_matrix = np.zeros(predictions.shape)

for i in range(len(predictions)):
    prediction_decision_function_matrix[i,np.argmax(predictions[i])] = 1
```

```python
from sklearn.metrics import multilabel_confusion_matrix
payload = multilabel_confusion_matrix(y_test,prediction_decision_function_matrix)
```

```python
fig, axes = plt.subplots(1, 3, sharey = True, figsize = (30,10))
sns.set(font_scale=2) # font size 2
plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=16)
sns.heatmap(payload[0], annot=True, ax = axes[0])
axes[0].set_title('setosa')
axes[0].set_xticklabels(['actual negative','actual positive']) 

sns.heatmap(payload[1], annot=True, ax = axes[1])
axes[1].set_title('versicolor')
axes[1].set_xticklabels(['actual negative','actual positive']) 
sns.heatmap(payload[2], annot=True, ax = axes[2])
axes[2].set_title('virginica')
axes[2].set_xticklabels(['actual negative','actual positive']) 
axes[0].set_yticklabels(['predicted negative','predicted positive']) 




plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction_decision_function_matrix))
```
